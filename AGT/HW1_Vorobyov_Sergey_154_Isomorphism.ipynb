#### Подгружаем необходимые библиотеки

```python
import networkx
from networkx.algorithms import isomorphism
import matplotlib.pyplot as plt
import pandas as pd
```

#### Нам необходимо работать с представлениями графов, определим конвертер из формата .csv в формат, удобный для чтения networkx

```python
#Строим граф для networkx
def graph_maker(A):
    graph = networkx.Graph()
    graph.add_nodes_from([i+1 for i in range(len(A))])
    for i in range(len(A)):
        for j in range(len(A[i])):
            if A[i][j] == 1:
                graph.add_edge(i+1,j+1)
    return graph

#Считываем .csv и преобразуем в список списков
def convert_from_csv_to_matrix(data):
    Matrix = []
    for i in range(len(data[0])):
        Matrix.append(list(data[i]))
    return Matrix
```

##### Подгружаем большую схему метро: метро Мадрида (117 станций)

```python
data = pd.read_csv('Madrid Scheme Vorobyov.csv', header = None)
data
```

```python
#Визуализируем полученный граф
%matplotlib inline
plt.figure(figsize = (15, 20))
G_1 = convert_from_csv_to_matrix(data)
G_1 = graph_maker(G_1)
networkx.draw(G_1, with_labels=True, font_weight='bold')
```

##### Подгружаем второе по величине большое метро: метро Гонг-Конга (79 станций)

```python
import igraph
#Конвертируем граф из networkx в igraph
def convert_from_VF2_to_igraph(G):
    H = igraph.Graph()
    for node in G:
        H.add_vertices(node)
    for (i,j) in G.edges():
        H.add_edges([(i,j)])
    return H

convert_from_VF2_to_igraph(G_1).get_edgelist()
```
